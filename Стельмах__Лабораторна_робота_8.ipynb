```python
df['Gender'].value_counts()
```

```python
df['Gender'] = df['Gender'].map({'Female': 0, 'Male': 1})
```

```python
df['Occupation'].value_counts()
```

```python
occupation_map = {
    'Nurse': 0,
    'Doctor': 1,
    'Engineer': 2,
    'Lawyer': 3,
    'Teacher': 4,
    'Accountant': 5,
    'Salesperson': 6,
    'Scientist': 7,
    'Software Engineer': 8,
    'Sales Representative': 9,
    'Manager': 10
}

df['Occupation'] = df['Occupation'].map(occupation_map)
```

```python
df['BMI Category'].value_counts()
```

```python
df['BMI Category'] = df['BMI Category'].replace({'Normal Weight': 'Normal'})
```

```python
df['BMI Category'].value_counts()
```

```python
bmi_map = {
    'Normal': 0,
    'Overweight': 1,
    'Obese': 2
}

df['BMI Category'] = df['BMI Category'].map(bmi_map)
```

```python
df['Sleep Disorder'].value_counts()
```

```python
df['Sleep Disorder'] = df['Sleep Disorder'].map({'Sleep Apnea': 0, 'Insomnia': 1})
```

```python
df['Blood Pressure'].value_counts()
```

```python
df[['Systolic', 'Diastolic']] = df['Blood Pressure'].str.split('/', expand=True).astype(int)
df.drop(columns=['Blood Pressure'], inplace=True)
```

```python
df['Pulse Pressure'] = df['Systolic'] - df['Diastolic']
df['Mean Arterial Pressure'] = (df['Systolic'] + 2 * df['Diastolic']) / 3
```

```python
df
```

```python
df.isnull().sum()
```

```python
most_common = df['Sleep Disorder'].mode()[0]
df['Sleep Disorder'] = df['Sleep Disorder'].fillna(most_common)
```

```python
df.isnull().sum()
```

```python
cor_matrix = df.corr()

plt.figure(figsize=(8, 5))
sns.heatmap(cor_matrix, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Матриця кореляції числових ознак")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['Stress Level'], df['Quality of Sleep'])
plt.xlabel('Рівень стресу')
plt.ylabel('Якість сну')
plt.title('Залежність між якістю сну та рівнем стресу (Matplotlib)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Stress Level', data=df, order=df['Stress Level'].value_counts().index)
plt.title('Розподіл рівня стресу (Stress Level)')
plt.xlabel('Рівень стресу')
plt.ylabel('Кількість')
plt.show()
```

```python
import plotly.express as px
fig = px.scatter(df, x='Age', y='Quality of Sleep', title='Залежність між віком та якістю сну')
fig.show()
```

```python
fig = px.box(df, x='Occupation', y='Quality of Sleep', title='Якість сну за професіями')
fig.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Sleep Duration'], bins=20, kde=True)
plt.title('Розподіл тривалості сну (Sleep Duration)')
plt.xlabel('Тривалість сну (години)')
plt.ylabel('Кількість')
plt.show()
```

```python
fig = px.scatter(df, x='Systolic', y='Diastolic', size='Stress Level', color='Sleep Disorder',
                 hover_data=['Sleep Duration'])
fig.update_layout(title='Тиск та стрес у людей з різними розладами сну')
fig.show()
```

```python
fig = px.box(df, x='Sleep Disorder', y='Heart Rate', color='Sleep Disorder')
fig.update_layout(title='Розподіл пульсу по типу порушення сну')
fig.show()
```

```python
sns.histplot(data=df, x='Sleep Duration', hue='Gender', multiple='stack')
plt.title('Розподіл сну за статтю')
plt.show()
```
