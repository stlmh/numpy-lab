```python
import plotly. express as px
```

```python
df = px.data.iris() # iris is a pandas DataFrame
df
```

```python
fig = px.scatter(df, x="sepal_width", y="sepal_length", color="species",
                 size="petal_length", hover_data=['petal_width'],
                 title="Sepal: Width vs Length",
                 color_discrete_sequence=px.colors.qualitative.Set2)
fig.update_layout(width=1000, height=800)
fig.show()
```

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="petal_length",
    color_continuous_scale="Viridis",
    title="Sepal Dimensions Colored by Petal Length"
)
fig.update_layout(template="plotly_white")
fig.show()
```

```python
fig = px.scatter(
    df,
    x="sepal_width",
    y="sepal_length",
    color="species",
    symbol="species",
    size="petal_length",
    hover_data=["petal_width"],
    title="Sepal Dimensions by Species"
)

fig.update_layout(
    template="plotly_white",
    width=1200,
    height=700,
    title_font_size=20
)

fig.show()
```

```python
df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", trendline="ols")
fig. show()
```

```python
df = px.data.gapminder().query("continent == 'Oceania'")
df.head()

```

```python
fig = px.line(df, x='lifeExp', y='gdpPercap', color='country')
fig.show()
```

```python
df = px.data.gapminder().query("year == 2007")

fig = px.scatter(
    df, x="gdpPercap", y="lifeExp",
    size="pop", color="continent",
    hover_name="country", log_x=True,
    title="GDP vs Life Expectancy (2007)",
    size_max=60,
    color_discrete_sequence=px.colors.qualitative.Set1
)
fig.show()
```

```python
df = px.data.gapminder().query("country == 'Japan'")

fig = px.line(
    df, x="year", y="lifeExp",
    title="Life Expectancy in Japan Over Time",
    line_shape='linear'
    )

fig.update_traces(line=dict(color=px.colors.qualitative.Pastel[3]))
fig.show()
```

```python
df = px.data.iris()

fig = px.violin(
    df, y="petal_length",x="species",
    color="species", box=True, points="all",
    title="Distribution of Petal Length by Species"
)
fig.show()
```

### Висновок

Під час роботи ми використовували бібліотеки **Plotly** та **NumPy**, щоб створити різноманітні графіки та виконати обчислення. Ми також використовували можливості зміни теми графіків, вибираючи між світлою або темною темою для більш комфортного перегляду даних. З допомогою бібліотеки Plotly можна легко адаптувати вигляд графіка до вимог користувача, що дуже важливо для створення презентаційних матеріалів.Щодо **NumPy**, то вона дозволяє зручно працювати з даними, генерувати числові масиви і виконувати математичні операції.
Таким чином, поєднання цих бібліотек дозволяє створювати не тільки красиві, але й інформативні графіки, що допомагає краще розуміти та аналізувати дані в різних сферах.
